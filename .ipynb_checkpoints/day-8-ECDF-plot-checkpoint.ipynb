# ECDF - Empirical Cumulative Distribution Functions

This partocular plot represents the proportion or count of observations falling below each unique value in a dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
file = pd.read_csv('supermarket_sales.csv')
df = file[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head()
```

```python
#ecdf plot
sns.ecdfplot(data=df,x='gross income')
```

```python
# hue : Semantic variable that is mapped to determine the color of plot elements.
sns.ecdfplot(data=df,x='gross income', hue='Gender')
```

```python
# stat : Distribution statistic to compute.
sns.ecdfplot(data=df,x='gross income', hue='Gender',stat='count')
```

```python
# change axis
sns.ecdfplot(data=df,y='gross income', hue='Gender')
```
