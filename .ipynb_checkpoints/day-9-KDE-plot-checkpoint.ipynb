# KDE Plot - Kernel Density Estimation Plot

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
file = pd.read_csv('supermarket_sales.csv')
df = file[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head(10)
```

### 1. Create a basic KDE plot for total column

```python
sns.kdeplot(data=df,x='Total')
```

### 2. Create KDE for all the numeric variables in dataframe

```python
sns.kdeplot(data=df)
```

### 3. Adjust the smoothing using bw_adjust

```python
sns.kdeplot(data=df,x='Total',bw_adjust=0.2)
```

### 4. Group the KDE on a category variable

```python
sns.kdeplot(data=df,x='Total',hue='Payment') # here payment is a categorical value
```

### 5. Stack KDE on a category using MULTIPLE argument

```python
sns.kdeplot(data=df,x='Total',hue='Payment',multiple='stack')
```

### 7. Use log scaling to map the variable in KDE

```python
sns.kdeplot(data=df,x='Total',log_scale=True)
```

### 7. Chage style of hued KDE using linewidth, palette, alpha etc

```python
sns.kdeplot(data=df,x='Total',hue='Payment',multiple='stack',linewidth=3,palette='Dark2',alpha=1)
```

### 8. Create a bivariate KDE

```python
sns.kdeplot(data=df,x='Unit price',y='gross income')
```

### 9. Group the bivariate KDE on a categorical variable and show the contours

```python
sns.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender')
```

```python
sns.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True)
```

```python
sns.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True,levels=5)
```

```python
sns.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True,levels=5,thresh=0.2)
```
