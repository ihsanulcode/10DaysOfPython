### What is series in pandas? 
A pandas Series is a one-dimensional array. It holds any data type supported in Python

```python
mySeries = [1,2,3,4,5]
s = pd.Series(mySeries)
```

```python
print(s)
```

```python
#print specific value
s[2]
```

```python
#lables using the index argument
myIndex = ['A','B','C','D','E']
data = pd.Series(mySeries,myIndex)
print(data)
```

```python
data['C']
```

### Series using Dictionary

```python
cars = {
    "Tesla":12,
    "Mercedes":42,
    "BMW":3
}

myCars = pd.Series(cars)
myCars
```

```python
myCars["BMW"]
```

### What is dataframe in pandas?
A dataframe is a 2D, size-mutable tabular data

```python
#importing numpy for creating dummy array data
import numpy as np
from numpy.random import randn
```

```python
myData = randn(5,5) #5 row 5 col
myRows = ['A','B','C','D','E']
myCols = ['Num1','Num2','Num3','Num4','Num5']
```

```python
#create a dataframe
df = pd.DataFrame(myData,myRows,myCols)
df
```

### Fix incomplete data

```python
#create a dummy data
stuff = {
    'A' : [1,2,3],
    'B' : [4,np.nan,6],
    'C' : [7,8,9],
    'D' : [10,11,12]
}
df = pd.DataFrame(stuff)
df
```

```python
# drop rows with null data
df.dropna()
```

```python
# drop col with null data
df.dropna(axis=1)
```

```python
# set the threshold
df.dropna(thresh=2,axis=1)
```

```python
# replace data with nan
df.fillna(value=10)
```

```python
df.fillna(value=df['B'].mean())
```

```python
df.fillna(value=df['B'].min())
```

```python
df.fillna(value=df['B'].max())
```

```python
df.fillna(value=df['B'].sum())
```
