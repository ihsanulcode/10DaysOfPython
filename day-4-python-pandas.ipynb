### What is series in pandas? 
A pandas Series is a one-dimensional array. It holds any data type supported in Python

```python
mySeries = [1,2,3,4,5]
s = pd.Series(mySeries)
```

```python
print(s)
```

```python
#print specific value
s[2]
```

```python
#lables using the index argument
myIndex = ['A','B','C','D','E']
data = pd.Series(mySeries,myIndex)
print(data)
```

```python
data['C']
```

### Series using Dictionary

```python
cars = {
    "Tesla":12,
    "Mercedes":42,
    "BMW":3
}

myCars = pd.Series(cars)
myCars
```

```python
myCars["BMW"]
```

### What is dataframe in pandas?
A dataframe is a 2D, size-mutable tabular data

```python
#importing numpy for creating dummy array data
import numpy as np
from numpy.random import randn
```

```python
myData = randn(5,5) #5 row 5 col
myRows = ['A','B','C','D','E']
myCols = ['Num1','Num2','Num3','Num4','Num5']
```

```python
#create a dataframe
df = pd.DataFrame(myData,myRows,myCols)
df
```

### Count Data

```python
dataque.Status.value_counts() #descending
```

```python
dataque.Status.value_counts(ascending=True) #ascending
```

```python
#Nan values
dataque.value_counts()
```

```python
dataque.value_counts(dropna=False)
```

```python
#get relative frequency - percentage
dataque.value_counts(dropna=False,normalize=True)*100
```

```python
# get specific item count
dataque.Status.value_counts().Completed
```

```python
# get specific item count
dataque.Status.value_counts()['Have Due']
```

```python
# count unique values - size
dataque.groupby('Status').size()
```

```python
# count unique values - count
dataque.groupby('Status').count()
```

```python
# get a count of all columns across all columns
dataque.apply(pd.value_counts)
```

### Add new columns

```python
# create fake data
gender = ['Male','Female']*8
gender.append('Male')

# Add col from list
dataque["Gender"] = gender
dataque
```

```python
# add default values
dataque['Enrolled'] = [True]*len(dataque)
dataque
```

```python
# Add nan value
dataque['Address'] = [np.nan]*len(dataque)
dataque
```

```python
# add col in position
dataque.insert(3,"Course",['DSA']*len(dataque),True)
dataque
```

```python
# add col and create new dataframe
#keep original 
newdf = dataque.assign(Batch=[3]*len(dataque))
newdf
```
