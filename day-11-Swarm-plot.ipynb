# Swarm Plot

We draw swarm plot to represent each of the data points including duplicates on the plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('supermarket_sales.csv')
df.head()
```

### Create a swarm plot on total column

```python
sns.swarmplot(data=df,y='Total')
```

```python
sns.swarmplot(data=df,y='Total')
plt.grid()
```

### Create swarm plot group by categories

```python
sns.swarmplot(data=df,x='Payment',y='Total')
```

```python
sns.swarmplot(data=df,x='Payment',y='Total',hue='Gender')

#In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding.
```

### Showing hue separatly on the categoriecal axis 

```python
sns.swarmplot(data=df,x='Payment',y='Total',hue='Gender',split=True)
```

### Styling a swarm change the marker size, color

```python
# change the marker
sns.swarmplot(data=df,x='Payment',y='Total',marker='*')
```

```python
# change the marker size
sns.swarmplot(data=df,x='Payment',y='Total',marker='*',size=8)
```

```python
# change the figure size for a cosure view
plt.figure(figsize=(20,5))

sns.swarmplot(data=df,x='Payment',y='Total',marker='*',size=8)
```

```python
# change the figure size for a cosure view
plt.figure(figsize=(20,5))

# Change the color of the marker
sns.swarmplot(data=df,x='Payment',y='Total',marker='*',size=8,color='red')
```

### Overlay a swarm plor on a box plot

```python
sns.boxplot(data=df,x='Payment',y='Total')
sns.swarmplot(data=df,x='Payment',y='Total')
```
