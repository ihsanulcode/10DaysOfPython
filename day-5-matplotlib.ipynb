# Matplotlib - Visualization with Python
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
from matplotlib import pyplot as plt
```

## 2D Plot

```python
x = [1,2,3,4,5]
y = [11,12,20,10,9]
```

```python
plt.plot(x,y,color='red')
```

```python
plt.plot(x,y,color='green',marker='+')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='--')
```

```python
plt.plot(x,y,alpha=0.5)
```

## Axis labels, Legend Grid

```python
days = [1,2,3,4,5,6,7]
maxt = [50,51,52,48,47,49,46]
mint = [43,42,40,44,33,35,37]
avgt = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,maxt,label="Max")
plt.plot(days,mint,label="Min")
plt.plot(days,avgt,label="Average")
plt.legend()
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,maxt,label="Max")
plt.plot(days,mint,label="Min")
plt.plot(days,avgt,label="Average")
plt.legend(loc='upper right')
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,maxt,label="Max")
plt.plot(days,mint,label="Min")
plt.plot(days,avgt,label="Average")
plt.legend(loc='best')
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,maxt,label="Max")
plt.plot(days,mint,label="Min")
plt.plot(days,avgt,label="Average")
plt.legend(loc='best', shadow=True)
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,maxt,label="Max")
plt.plot(days,mint,label="Min")
plt.plot(days,avgt,label="Average")
plt.legend(loc='upper right')
plt.grid()
```

## Bar Chart

```python
company = ['Google','Amazon','Tesla','Meta']
revenue = [90,134,56,27]
profit = [40,20,34,12]
```

```python
plt.bar(company,revenue,label='Revenue')
plt.xlabel('Company Name')
plt.ylabel('Generated Revenue')
plt.title('US Company States')
plt.legend()
```

```python
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.xlabel('Company Name')
plt.ylabel('Generated Revenue')
plt.title('US Company States')
plt.legend()
```
